```python
import requests

# URL do endpoint (ajuste conforme o ambiente)
BASE_URL = 'http://localhost:3000'  # Vite proxy ou 5173 se direto

# 1. Obter id_encontro
response = requests.get(f'{BASE_URL}/api/encontros/abertos')
if response.status_code == 200:
    encontros = response.json()
    if encontros:
        id_encontro = encontros[0]['id']
        print(f'id_encontro utilizado: {id_encontro}')
    else:
        print('Nenhum encontro aberto encontrado')
else:
    print(f'Erro ao obter encontros: {response.status_code}')
```

```python
# 2. Testar POST /api/inscricoes/create
payload = {
    'id_encontro': id_encontro,
    'nome_adolescente': 'Teste Checklist',
    'data_nascimento': '2010-01-01',
    'telefone_adolescente': '11987654321',
    'nome_responsavel': 'Responsavel Teste',
    'telefone_responsavel': '11876543210',
    'aceite_termos': True,
    'bairro': 'Bairro Teste',
    'paroquia': 'Paroquia Teste',
    'participou_antes': False,
    'observacoes': 'Teste do checklist'
}

response = requests.post(f'{BASE_URL}/api/inscricoes/create', json=payload)
print(f'Status HTTP: {response.status_code}')
print(f'Body retornado: {response.json()}')

if response.status_code == 201:
    inscricao_id = response.json()['data']['id']
    print(f'Inscrição criada com ID: {inscricao_id}')
elif response.status_code == 200 and response.json().get('duplicate'):
    print('Duplicata detectada')
else:
    print('Erro na criação')
```
